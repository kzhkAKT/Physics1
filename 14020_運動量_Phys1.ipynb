# 力積と運動量

\begin{align}
& m v' - m v = F \Delta t&
\end{align}

## 運動量 $p$

\begin{align*}
 p = m v
\end{align*}

> 運動量 [kg m/s]=質量 [kg] × 速度 [m/s]

速度も **瞬間の速度** なので, 観測したその瞬間の情報を利用して計算すれば良い.
