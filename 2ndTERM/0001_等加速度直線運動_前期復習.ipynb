# 位置 $x$ と変位 $\Delta x$

前期では一次元の運動を考えてきました.
- **位置 $x$**: 原点を決め, 特定の方向を **正** とすることで決まる
- **変位 $\Delta x$**: 始め観察した位置 $x_1$から,    
  後の位置 $x_2$ がどれだけ移動したかを表す量 **$\Delta = x_2 - x_1$**

# 速度 $v$, 加速度$a$

## 平均の速度 $\overline{v}$

\begin{align*}
\overline{v}&=\dfrac{\varDelta x}{\varDelta t}
=\dfrac{x_2 - x_1}{t_2 - t_1}
\end{align*}

## 平均の加速度$\overline{a}$

\begin{align*}
\overline{a}
&=\dfrac{\varDelta v}{\varDelta t}
=\dfrac{v_2 - v_1}{t_2 - t_1}
\end{align*}

## 瞬間の速度 $v$, 瞬間の加速度$a$
\begin{align*}
v 
&= \lim_{\Delta t \to 0} \dfrac{\varDelta x}{\varDelta t}
= \dfrac{{\rm d} x}{{\rm d}t}\\[2mm]  
a
&= \lim_{\Delta t \to 0} \dfrac{\varDelta v}{\varDelta t}
= \dfrac{{\rm d} v}{{\rm d}t}\\[2mm]  
\end{align*}

一般的には
- 瞬間の速度を **速度**
  - $x-t$ グラフの接線の傾きが **速度**
- 瞬間の加速度を **加速度**
  - $v-t$ グラフの接線の傾きが **加速度**

と表現されます.

## グラフの面積と物理量

### 速度から変位へ

$\require{cancel}$
\begin{align*}
\overline{v}&=\dfrac{\varDelta x}{\varDelta t}\\
\overline{v}{\varDelta t} 
&=\dfrac{\varDelta x}{\bcancel{\varDelta t}} {\bcancel{\varDelta t}}\\
\Delta x &= \overline{v}{\varDelta t}
\end{align*}
> $v-t$ グラフにおいて, ($t$軸で)囲まれる面積は変位 $\Delta x$ を表す

これを拡張して

>> $v-t$ グラフにおいて, ($t$軸で)囲まれる面積は移動距離を表す

ということができます.

### 加速度から速度変化へ


\begin{align*}
\overline{a}&=\dfrac{\varDelta v}{\varDelta t}\\
\overline{a}{\varDelta t} 
&=\dfrac{\varDelta v}{\bcancel{\varDelta t}} {\bcancel{\varDelta t}}\\
\Delta v &= \overline{a}{\varDelta t}
\end{align*}
> $a-t$ グラフにおいて, ($t$軸で)囲まれる面積は速度変化 $\Delta v$ を表す

# 等加速度直線運動

$a$: ずっと一定, 1方向

## $t$ 秒後の速度 $v$

$\require{cancel}$
\begin{align*}
a
&= \overline{a} 
=\dfrac{v_2 - v_1}{t_2 - t_1}
=\dfrac{v - v_0 }{t - 0}\\[1mm]
v
&= v_0 + a t
\end{align*}

## $t$ 秒後の原点からの変位 $s$

$v-t$ グラフを書きその面積を求めると
\begin{align*}
s
&= v_0 \cdot t + \dfrac{1}{2}a t \cdot t\\[1mm]
&= v_0 t + \dfrac{1}{2}a t^2
\end{align*}

## 時間依存の消去

等加速度直線運動では $v,s$ とも $t$ による媒介変数表示
\begin{cases}
v
&= v_0 + a t\\[1mm]
s
&= v_0 t + \dfrac{1}{2}a t^2,
\end{cases}

時間の変数 $t$ を消去すると
\begin{align*}
v^2 - v_0^2 &= 2 a s
\end{align*}
が得られる.
