### 仕事率

仕事率 (Power, $P$)
> 単位時間当たりにした仕事

力が物体にした仕事を $W$ とし, その仕事を行うのに $t$ 秒かかったとすると  
仕事率は
\begin{align*}
P =\dfrac{W}{t}
\end{align*}

- 仕事率の単位(ワット):: $\quad {\rm W}$
  - 仕事率の定義に従うと: $\qquad 1 {\rm W} = \dfrac{1 {\rm J}}{ 1 {\rm s}}$  
    - 仕事の単位の定義を利用し, 等速で物体を動かしたとすると:   
    $\qquad 1 {\rm W} = \dfrac{1 {\rm J}}{ 1 {\rm s}} = \dfrac{1 {\rm N} \cdot 1 {\rm m} }{ 1 {\rm s}} = 1 {\rm N} \cdot 1 {\rm m/s} $
