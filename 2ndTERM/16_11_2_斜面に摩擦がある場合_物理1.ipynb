# 平面空間での運動

## 斜面上にある物体の運動

### 斜面に摩擦がある場合

#### 摩擦のある斜面を滑る物体に働く力
<div align="center">
<img src="https://docs.google.com/drawings/d/e/2PACX-1vQxOwbR76v-_jD3QEO7RezhAdEFLxMW_dQ7ViOUWxNZVsuP3hhzxtSeYduYN6ZxQNfwD531lvb8UCUH/pub?w=480">
</div>

水平方向に対して, 傾角(けいかく) $\theta$ の **摩擦がある** 斜面を滑る  
物体の運動方程式を考えます.

出発点は, 瞬間を捉えることができる, 運動方程式
\begin{align}
m \vec{a} = \vec{F}
\end{align}

$\require{color}$
物体の運動に合わせて
- 斜面下向きを $x$軸の _正_
- 斜面に対して垂直な方向を $y$軸の _正_

>> 座標の取り方が、これまでと違っていることに注意してください.

また,
- 物体が動摩擦力 $f'$  を受けながら滑る
  - 動摩擦力の大きさは $f'=\mu' N$

として, 成分表示をすると

重力ベクトル $\vec{W}$, 垂直抗力ベクトル $\vec{N}$, 動摩擦力 $\vec{f'}$ は

\begin{align}
\vec{W}
  =
   \left(\begin{array}{c}
    W_x \\
    W_y
  \end{array}
  \right)
  =
   \left(\begin{array}{c}
    {+mg \sin\theta} \\
    {-mg \cos \theta}
  \end{array}
  \right),
\qquad
\vec{N} =   \left(\begin{array}{c}
    0 \\
    + N
  \end{array}
\right),
\qquad
\vec{f'} =   \left(\begin{array}{c}
    \color{purple}{-f'} \\
    0
  \end{array}
\right),
\end{align}


加速度ベクトル $\vec{a}$ もこの座標系を用いて
\begin{align}
\vec{a} =   \left(\begin{array}{c}
    a_x \\
    a_y
  \end{array}
\right),
\end{align}
とし,

 運動方程式のそれぞれの成分を表現すると
\begin{align}
m \vec{a} = \vec{F} = \vec{N} + \vec{W} + \color{purple}{\vec{f'}}:
\quad
\begin{cases}
m a_x &= 0 + mg {\sin\theta} + \left(\color{purple}{-f'} \right)\\
m a_y &= N + \left( - mg\cos\theta \right) + 0
\end{cases}
\quad
\to
\quad
\begin{cases}
m a_x &= {mg \sin\theta} + \left(\color{purple}{-f'} \right)\\
m a_y &= N + (- mg\cos\theta)
\end{cases}
\end{align}

- 斜面を滑り落ちるときは, 物体は斜面に垂直な方向($y$軸)には  
  変位を生じない[飛んだり、跳ねたりしない]ので, $a_y = 0$


$\require{cancel}$
\begin{align}
\begin{cases}
{m} a_x &= {{m}g \sin\theta} + \left(\color{purple}{-f'} \right)\\
m \cancelto{0}{a_y}
 &= N {- mg\cos\theta}
\end{cases}
\quad
\to
\quad
\begin{cases}
{m} a_x &= {{m}g \sin\theta} + \left(\color{purple}{-f'} \right)\\
 N &= { mg\cos\theta}
\end{cases}
\end{align}

ここで動摩擦力の大きさ $\color{red}{f'}$ へ, 垂直抗力 $N$ を代入し, 整理すると
\begin{align*}
m a_x &= m g \sin \theta - \mu' N \\[1mm]
&= m g \sin \theta - \mu' m g \cos \theta \\[2mm]
\bcancel{m} a_x &= \bcancel{m} g \left( \sin \theta - \mu' \cos \theta \right)\\[3mm]
a_x &= g \left( \sin \theta - \mu' \cos \theta \right)
\end{align*}

- 斜面方向 $x$ の加速度は, 斜面の角度によって決まり, 質量 $m$ によらない
  - 動摩擦力の影響を受け, この加速度の大きさは, 摩擦のないときより _小さくなる_
- 垂直抗力は, 力のつり合いから, $mg\cos\theta$
