# 平面空間での運動

## 斜面上にある物体の運動

### 斜面に摩擦がない場合

```
#@title なめらかな斜面を滑る物体に働く力
#https://en.wikipedia.org/wiki/Frictionless_plane
%%svg
https://upload.wikimedia.org/wikipedia/commons/9/9e/Free_body_frictionless.svg
```

水平方向に対して, 傾角(けいかく) $\theta$ の *なめらかな* 斜面滑る  
物体の運動方程式を考えます.

出発点は, 瞬間を捉えることができる, 運動方程式
\begin{align}
m \vec{a} = \vec{F}
\end{align}

$\require{color}$
物体の運動に合わせて
- 斜面下向きを $x$軸の _正_
- 斜面に対して垂直な方向を $y$軸の _正_

>> 座標の取り方が、これまでと違っていることに注意してください.

として, 成分表示をすると

重力ベクトル $\vec{W}$ や 垂直抗力ベクトル $\vec{N}$ は

\begin{align}
\vec{W}
  =
   \left(\begin{array}{c}
    W_x \\
    W_y
  \end{array}
  \right)
  =
   \left(\begin{array}{c}
    \textcolor{red}{+mg \sin\theta} \\
    \textcolor{blue}{-mg \cos \theta}
  \end{array}
  \right),
\qquad
\vec{N} =   \left(\begin{array}{c}
    0 \\
    + N
  \end{array}
\right),
\end{align}


加速度ベクトル $\vec{a}$ もこの座標系を用いて
\begin{align}
\vec{a} =   \left(\begin{array}{c}
    a_x \\
    a_y
  \end{array}
\right),
\end{align}
とし,

 運動方程式のそれぞれの成分を表現すると
\begin{align}
m \vec{a} = \vec{F} = \vec{N} + \vec{W}:
\quad
\begin{cases}
m a_x &= 0 \textcolor{red}{+mg \sin\theta}\\
m a_y &= N + \left( \textcolor{blue}{- mg\cos\theta} \right)
\end{cases}
\quad
\to
\quad
\begin{cases}
m a_x &= \textcolor{red}{+mg \sin\theta}\\
m a_y &= N \textcolor{blue}{- mg\cos\theta}
\end{cases}
\end{align}

- 斜面を滑り落ちるときは, 物体は斜面に垂直な方向($y$軸)には  
  変位を生じない[飛んだり、跳ねたりしない]ので, $a_y = 0$
 

$\require{cancel}$
\begin{align}
\begin{cases}
\bcancel{m} a_x &= \textcolor{red}{+\bcancel{m}g \sin\theta}\\
m \cancelto{0}{a_y}
 &= N \textcolor{blue}{- mg\cos\theta}
\end{cases}
\quad
\to
\quad
\begin{cases}
a_x &= \textcolor{red}{g \sin\theta}\\
 N &= \textcolor{blue}{ mg\cos\theta}
\end{cases}
\end{align}

- 斜面方向 $x$ の加速度は, 斜面の角度によって決まり, 質量 $m$ によらない
- 垂直抗力は, 力のつり合いから, $mg\cos\theta$
