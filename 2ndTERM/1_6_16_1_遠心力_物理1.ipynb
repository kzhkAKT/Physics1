上の図のように等速円運動する物体について
1. 慣性系:: 回転する円盤の外から、静止した状態で、物体を観測する
2. 非慣性系:: 回転する円盤に乗って、円盤と一緒に回転しながら、物体を観測する

ことを考えていきます。

### 慣性系から等速円運動する物体を観察

慣性系は、これまで学んできた円盤外部からの(静止した状態での)観測なので  
物体には 常に中心向きの力 **向心力** が働いていると捉えることができます。

この向心力の大きさは、等速円運動の関係より
\begin{align}
&\begin{cases}
v&= r \omega\\
a&= v \omega\\
\end{cases}\\
F
&= m a = m v\omega = m \left( r\omega \right)\omega \\
&= m r \omega^2
\end{align}




### 非慣性系から等速円運動する物体を観察

回転する円盤に乗って、物体を観測する場合は  
観測者は中心向きの加速度 a を受けながら観測しています。  
つまり、物体には **慣性力** が働きます

特に、等速円運動する物体に対し、_中心と物体を結ぶ直線上_ に働く 慣性力 を
>> **遠心力** と言います

遠心力は、
1. 非慣性系(円盤に乗った観測者)から見て、中心に対して外向き
2. その大きさは $\quad f=ma=mr\omega^2$
