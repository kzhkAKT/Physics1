## 等速円運動

uniform circular motion

- 一定の半径 $r$ の円周上を一定の速さ $v$ で 回転している質量 $m$ の質点の運動
  - 一周($360^{\circ}, 2\pi$)すると元の位置に戻ります
    - 物体が一周する時間(${\rm \ [s]}$)を **周期** $T$
      - 周期 $T$ の逆数 $\dfrac{1}{T}$ を **回転数** $n$   
    - 周期 $T$ と 回転数 $n$ の 関係
    $: \qquad {\rm 1 \ [s] : T \ [s]= n\ (周): 1 \ (周) }$

> Q. 回転数の単位を求めてください. (答え: ${\rm [1/s]}$ )

### 等速円運動物体の速さ $v$

速度は変位と時間変化の比で定義され,  
その大きさである **速さ** は
\begin{align}
\left| \vec{v} \right| = v = \dfrac{L_{r}}{T} = \dfrac{2\pi r}{T}
\end{align}

### 等速円運動物体の加速度の大きさ $a$

加速度は速度の変化と時間変化の比で定義され,  
その大きさは, ホドグラフを参考にし
\begin{align}
\left| \vec{a} \right| = a = \dfrac{L_{v}}{T} = \dfrac{2\pi v}{T}.
\end{align}

これを, 等速円運動する物体の速度 $v$ の関係を用いて整理すると

\begin{align}
a &= \dfrac{2\pi v}{T} =  \dfrac{2\pi r}{T} \dfrac{v}{r} \
= v \dfrac{v}{r}  \\[3mm]
a &= \dfrac{v^2}{r}
\end{align}

#### 等速円運動する物体の加速度の方向

- 等速円運動する物体の, ホドグラフの接線方向
  - その瞬間の速度 $\vec{v}$ と $\dfrac{\pi}{2}, 90^{\circ}$ で交わる
    - 等速円運動の円軌道で考えれば, **常に中心を向く**

### 向心力

加速度 $\vec{a}$ がわかったので, 等速円運動を行う  
物体に働く力 $\vec{F}$ を調べます.

\begin{align}
m \vec{a} &=  \vec{F}\\[2mm]
F &= \left| \vec{F} \right| = \left| m \vec{a} \right| = m \left| \vec{a} \right| \\[1mm]
F &= ma = m \dfrac{v^2}{r}
\end{align}

> $\vec{F}$ は運動方程式から加速度 $\vec{a}$ に平行なので,  
 _常に中心をむくため_, **向心力** と呼ばれます.

#### 角速度 $\omega$ の最も簡単な定義

>> 単位時間 (${\rm 1\ [s]}$) あたりの 角度変化 (${\rm  rad}$)

> Q. 等速円運動している物体が, 時間 $t$ の間に,   
角度が $\theta$ 変化した. この時の角速度 $\omega$ を求めよ.

##### 角速度 $\omega$ と周期の関係

角速度の定義と, 等速円運動の性質より次の関係が成り立ちます
\begin{align}
\omega = \dfrac{\theta}{t} =  \dfrac{2\pi}{T},
\end{align}
ここで $T$ は周期で, 一周($2\pi$)するための時間です.
