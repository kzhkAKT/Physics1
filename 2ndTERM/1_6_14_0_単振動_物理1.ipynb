#### 等速円運動の速度, 加速度, 向心力

周期 $T$ と回転数 $n$

- 周期 $T$ の単位は，時間なので $\quad {\rm s}$
- 回転数の単位は，時間の **逆数** なので $\quad {\rm 1/s}$

\begin{align}
1:n &= T:1\\
  nT & = 1\\[2mm]
  n &= \dfrac{1}{T}
\end{align}

角速度 $\omega$
\begin{align}
\omega
&= \dfrac{\theta}{t} &= \dfrac{2\pi}{T}\\
&= 2\pi \ \dfrac{1}{T} &= 2\pi n\\
\end{align}


$\require{cancel}$
大きさ: 等速円運動の速度, 加速度, 向心力
\begin{align}
v &= \dfrac{2\pi r}{T} = r \cancelto{\omega}{\dfrac{2\pi}{T}}=r \omega\\[2mm]
a &= \dfrac{2\pi v}{T} = v \cancelto{\omega}{\dfrac{2\pi}{T}}= v \omega = r \omega^2\\[3mm]
F &= ma = m v \omega 
=
\begin{cases}
 m \left( r \omega \right) \omega & =mr \omega^2\\[1mm]
 m v \left( \dfrac{v}{r} \right)&= m \dfrac{v^2}{r}
\end{cases}
\end{align}

振動数 $f:{\rm \quad Hz}$, 角振動数 $f:{\rm \quad rad/s}$


#### 単振動の変位


![https://www.ikstudie.com/articles/442](https://s3-ap-northeast-1.amazonaws.com/ikstudie.com/uploads/ckeditor/pictures/288/content___.png)
---
[出典:受験の月](https://examist.jp/physics/mechanics/tansindou/)

時刻 $t$ での，**位相**(初めの点からの角度$\theta$)とすると,  
半径 $A$ を射影した $x$ は,

\begin{align}
x &= A \sin \theta
\end{align}

角速度の定義より $\omega = \frac{\theta}{t}$ なので，  
(等速円運動では$\omega = \frac{\theta}{t}=\frac{2\pi}{T}$)

\begin{align}
x
&= A \sin \omega t\\[2mm]
&=  A \sin \dfrac{2\pi}{T} t\\[1mm]
&=  A \sin {2\pi}f t\\[1mm]
\end{align}


#### 単振動の速度

![https://www.ikstudie.com/articles/442](https://s3-ap-northeast-1.amazonaws.com/ikstudie.com/uploads/ckeditor/pictures/288/content___.png)
---
[出典:受験の月](https://examist.jp/physics/mechanics/tansindou/)

時刻 $t$ での，**位相**(初めの点からの角度$\theta$)とすると,  
等速円運動の速さ $A\omega$ を射影した $v_x$ は,

\begin{align}
v_x
&= A \omega\sin \left (\theta + \dfrac{\pi}{2} \right)\\[2mm]
&= A \omega\cos \theta 
\end{align}

角速度の定義より $\omega = \frac{\theta}{t}$ なので，  
(等速円運動では$\omega = \frac{\theta}{t}=\frac{2\pi}{T}$)

\begin{align}
v_x &= A \cos \omega t
\end{align}


#### 単振動の加速度

![https://www.ikstudie.com/articles/442](https://s3-ap-northeast-1.amazonaws.com/ikstudie.com/uploads/ckeditor/pictures/288/content___.png)
---
[出典:受験の月](https://examist.jp/physics/mechanics/tansindou/)

時刻 $t$ での，**位相**(初めの点からの角度$\theta$)とすると,  
等速円運動の大きさ $A\omega^2$ を射影した $a_x$ は,

\begin{align}
a_x
&= A \omega^2\sin \left (\theta + \pi \right)\\[2mm]
&= - A \omega^2\sin \theta 
\end{align}

角速度の定義より $\omega = \frac{\theta}{t}$ なので，  
(等速円運動では$\omega = \frac{\theta}{t}=\frac{2\pi}{T}$)

\begin{align}
a_x &= - A \omega^2 \sin \omega t
\end{align}


## 単振動のまとめ

![https://www.ikstudie.com/articles/442](https://s3-ap-northeast-1.amazonaws.com/ikstudie.com/uploads/ckeditor/pictures/288/content___.png)
---
[出典:受験の月](https://examist.jp/physics/mechanics/tansindou/)

周期$T$と振動数の関係
\begin{align}
f \ T =1
\end{align}

角振動数(=角速度)
\begin{align}
\omega = 2\pi f
\end{align}

単振動の速度，速度，加速度
\begin{cases}
x &= A \sin \omega t\\[1mm]
v_x &= A \omega \cos \omega t\\[1mm]
a_x &= -A \omega^2 \sin \omega t
\end{cases}
