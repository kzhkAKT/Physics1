## 単位


5 kg = 5 $\times$ 1 kg &ensp;
"1 kg" の部分を単位(unit)と呼ばれます.   
この場合, **1 kg** は **質量の単位** です.

### 量と数の違い
* 量は、数値と単位との積で成り立ちます.

* 同じ量(単位)のものは, 足し算と引き算(加減)ができます.
    * 異なる量(単位)のものは、単純に加減ができません.
* 理工学系で利用する **(組立)単位** は, *(基本)単位* の掛け算と割り算(乗除)で組み立てられます.
    * 例: m/s は 長さの単位 1 m と時間の単位 1 s の割り算で組まれています.
* 理工学系では等式の両辺は同じ量/単位をもちます.
    * $m\cdot a = F$
      &ensp; 前期中盤で紹介する **力の単位 N** の定義
    * $W = F \cdot x$
      &ensp; 後期に紹介する **仕事の単位 J** の定義

* 異なる単位は等式で結びつけることは <font color="red">*できません*</font>.
    * $5\ {\rm kg} \neq 5\ {\rm m}$

* 量(=数値$\times$単位) を **物理量** と呼ぶことがあります.

## 物理量と量の計算

物理量 $A$ は、次のように「数値」と「単位」の掛け算として表されます。

$$
A = \{A\}[A] = \{A\} \times [A], \qquad
\begin{cases}
\{A\}: \text{数値 (numerical value)}\\
[A]: \text{単位 (unit)}
\end{cases}
$$

---

参考: [IUPAC グリーンブック 1.1「物理量と量の計算」](https://www.ir.ethz.ch/GreenBook/GBAV_as_appeared_2023_2024.pdf#page=13.16)

### 前提条件: 次元の整合性 (単位の整合性)
物理量を足したり引いたりするためには、**それらが全く同じ単位（次元）を持っていなければなりません**。

ここでは、それぞれの単位が同じであると仮定します：$[A_1] = [A_2] = [A]$



### 基本的な足し算と引き算
#### 1. 足し算 (加算)
同じ単位を持つ2つの物理量を足し合わせる場合、単位 $[A]$ はそのまま共通のくくりとして残り、数値 $\{A_1\}$ と $\{A_2\}$ のみが足し合わされます。

$$
\require{cancel}
\begin{align*}
A_1 + A_2 &= \{A_1\}[A_{\cancel{1}}] + \{A_2\}[A_{\cancel{2}}] \\
&= (\{A_1\} + \{A_2\})[A]
\end{align*}
$$

> **具体例 (Example):**
> 2つの荷物があり、それぞれの質量が $m_1 = 12 \text{ kg}$ と $m_2 = 5 \text{ kg}$ のとき、全体の質量 $M$ を求める場合
> $$M = m_1 + m_2 = (12 + 5)[\text{kg}] = 17 \text{ kg}$$
> 単位の $\text{kg}$ はそのまま共通のくくりとして残り、数値同士が足し合わされます。

#### 2. 引き算 (減法または差分)
足し算と同様に、物理量の引き算をする場合も単位 $[A]$ は変わらず、数値同士のみを引き算します。

$$
\begin{align*}
A_1 - A_2 &= \{A_1\}[A] - \{A_2\}[A] \\
&= (\{A_1\} - \{A_2\})[A]
\end{align*}
$$

> **具体例 (Example):**
> 実験前の温度が $T_1 = 298 \text{ K}$、実験後の温度が $T_2 = 350 \text{ K}$ のとき、温度変化（差） $\Delta T$ を求める場合
> $$\Delta T = T_2 - T_1 = (350 - 298)[\text{K}] = 52 \text{ K}$$
> こちらも単位の $\text{K}$ (ケルビン) はそのまま残り、数値同士の引き算になります。

### 掛け算と割り算

足し算や引き算の「同じ単位でなければ計算できない」というルールとは異なり、物理量の掛け算と割り算では**異なる単位同士**を計算することができます。
このとき、数値だけでなく**単位同士も掛けたり割ったりする**ため、結果としてまったく新しい単位（次元）を持つ物理量が生まれます。

#### 3. 掛け算 (乗法)
2つの物理量 $A$ と $B$ を掛ける場合、数値は数値同士、単位は単位同士でそれぞれ掛け合わせます。

$$
\begin{align*}
A \times B &= (\{A\}[A]) \times (\{B\}[B]) \\
&= (\{A\} \times \{B\}) ([A] \times [B])
\end{align*}
$$

> **具体例 (Example):**
> 速さ $v = 5 \text{ m/s}$ と 時間 $t = 10 \text{ s}$ を掛けて、移動距離 $x$ を求める場合
> $$x = v \times t = (5 \times 10) (\text{m/s} \times \text{s}) = 50 \text{ m}$$
> 単位の $\text{s}$ (秒) が約分されて消え、新しく $\text{m}$ (メートル: 距離の単位) が導かれます。

#### 4. 割り算 (除法 / 商)
物理量 $A$ を $B$ で割る場合も同様に、数値は数値同士、単位は単位同士でそれぞれ割り算（分数）にします。

$$
\begin{align*}
\frac{A}{B} &= \frac{\{A\}[A]}{\{B\}[B]} \\
&= \left(\frac{\{A\}}{\{B\}}\right) \left(\frac{[A]}{[B]}\right)
\end{align*}
$$

> **具体例 (Example):**
> 質量 $m = 10 \text{ kg}$ を 体積 $V = 2 \text{ m}^3$ で割って、密度 $\rho$ (ロー) を求める場合
> $$\rho = \frac{m}{V} = \left(\frac{10}{2}\right) \left(\frac{\text{kg}}{\text{m}^3}\right) = 5 \text{ kg/m}^3$$
> 異なる単位が組み合わさり、そのまま $\text{kg/m}^3$ (キログラム毎立方メートル: 密度の単位) という新しい次元として表現されます。

### ⚠️ NG例：やってはいけない計算

物理量の足し算・引き算において、前提条件である **「単位（次元）が完全に一致していること」** を守らないと、物理的に意味を持たない式になったり、間違った結果を導いたりしてしまいます。特によくある間違いを2つ挙げます。

#### NG例 1: 異なる次元（種類）の物理量を足し引きする
長さと時間、質量と速さなど、物理的な性質（次元）が全く異なるものを直接足したり引いたりすることは絶対にできません。

$$
5 \text{ m} + 3 \text{ s} \neq 8 \text{ (何らかの単位)} \quad \text{（※計算不可）}
$$

> **解説:**
> 「5メートルの距離」に「3秒の時間」を足しても、現実の物理現象として意味を持ちません。（※掛け算や割り算であれば、異なる次元同士を計算して「速さ」などの新しい物理量を作ることができます）。

#### NG例 2: 同じ次元でも、単位（接頭語など）を揃えずに数値を計算する
どちらも同じ「質量」の物理量であっても、$\text{kg}$（キログラム）と $\text{g}$（グラム）のように単位の基準が異なるまま、数値だけを足し引きしてはいけません。

$$
1 \text{ kg} + 500 \text{ g} \neq 501 \text{ kg} \quad \text{（※誤り）}
$$

> **解説:**
> 単位を無視して数値の「1」と「500」だけを足してしまうのは、テスト等で非常によくあるミスです。この計算を正しく行うためには、**必ず計算前にどちらか一方の単位に変換して揃える**必要があります。
>
> **正しい計算の例:**
> * $\text{kg}$ に揃える場合: $1 \text{ kg} + 0.5 \text{ kg} = 1.5 \text{ kg}$
> * $\text{g}$ に揃える場合: $1000 \text{ g} + 500 \text{ g} = 1500 \text{ g}$

### ⚠️ 掛け算・割り算でのNG例（よくある間違い）

足し算・引き算と違い、掛け算・割り算は**異なる次元同士でも計算可能**です。しかし、「何でも適当に掛けて/割ってよい」わけではありません。以下のようなミスが多発するため注意が必要です。

#### NG例 1: 数値だけを計算して、単位の計算を忘れる（または勝手に省略する）
掛け算・割り算では、数値だけでなく**単位同士も掛け算・割り算される**ことを忘れてはいけません。

$$
10 \text{ m} \div 2 \text{ s} \neq 5 \text{ m} \quad \text{（※誤り）}
$$

> **解説:**
> 「$10 \div 2 = 5$」という数値の計算だけに満足してしまい、単位を元の $\text{m}$ のままにしてしまうミスです。正しくは単位も割って $5 \text{ m/s}$ としなければなりません。
> 力学などで「力のモーメント（$\text{N} \cdot \text{m}$）」を求めるときに、距離を掛けたのに単位を $\text{N}$（ニュートン）のままにしてしまうのも、これと同じミスです。

#### NG例 2: 接頭語（k, c, mなど）が混在したまま計算し、実用性のない単位を作ってしまう
面積や体積などを求める際、異なるスケールの単位をそのまま掛けると、後で直感的に理解しにくい（または公式に代入できない）単位になってしまいます。

$$
2 \text{ km} \times 3 \text{ cm} = 6 \text{ km} \cdot \text{ cm} \quad \text{（※数学的には可能だが、物理・工学では実質NG）}
$$

> **解説:**
> $\text{km} \cdot \text{cm}$ という単位は実用上ほとんど意味を成さず、その後の計算ミスの大きな原因になります。掛け算・割り算であっても、計算前に **基本となる単位（SI単位系の $\text{m}$ など）に変換して揃えてから計算する** のが工学の鉄則です。
>
> **正しいアプローチ:** $2000 \text{ m} \times 0.03 \text{ m} = 60 \text{ m}^2$

#### NG例 3: 関数（$\sin, \cos, e$ など）の「中身」に単位を残してしまう
高専の物理や専門科目で出てくる三角関数や指数関数の「中身（引数）」は、掛け算や割り算の結果として**必ず「無次元（単位を持たない状態）」**にならなければなりません。

$$
\sin(5 \text{ m} \times 2 \text{ s}) \Rightarrow \sin(10 \text{ m}\cdot\text{s}) \quad \text{（※絶対にNG！）}
$$

> **解説:**
> マクローリン展開などの数学的な性質上、「$10 \text{ m}\cdot\text{s}$ のサイン」というものは物理的に存在しません。もし計算途中で $\sin$ や $e$ の中身に単位が残ってしまったら、その時点で「計算ミスをしている」という強力なエラー検知センサーになります。



#### 【発展】対数関数（$\log, \ln$）の例外的な取り扱い

> 発展事項ですので、頭の片隅に残しておいてください

$\sin$ や $e^x$ では単位を残すことはできませんが、**対数関数は例外**です。対数関数の性質を使うと「数値」と「単位」を分離できるため、計算の途中で単位を含む式が成立します。

##### 対数による物理量 $A$ の変換

物理量 $A = \{A\}[A]$ の自然対数 $\ln A$ をとると、次のように展開できます。

$$
\begin{align*}
\ln A &= \ln (\{A\}[A]) \\
&= \ln \{A\} + \ln [A]
\end{align*}
$$

この性質を利用すると、以下のように物理量の対数同士の計算が可能になります。

**1. 引き算 (Subtraction / Difference)**

同じ次元を持つ物理量の対数の差をとると、単位の項が相殺（$\ln 1 = 0$）され、無次元の数値の対数だけが残ります。

$$
\require{cancel}
\begin{align*}
\ln A_1 - \ln A_2 &= \ln (\{A_1\}[A_{\cancel{1}}]) - \ln (\{A_2\}[A_{\cancel{2}}]) \\
&= \ln (\{A_1\}/\{A_2\}) + \cancelto{0}{\ln ([A]/[A])}
\end{align*}
$$
または、分数の形にまとめてから約分することもできます。
$$
\begin{align*}
\ln A_1 - \ln A_2 &= \ln \left(\frac{A_1}{A_2}\right) \\
&= \ln \frac{\{A_1\}[A_{\cancel{1}}]}{\{A_2\}[A_{\cancel{2}}]} \\
&= \ln (\{A_1\} / \{A_2\})
\end{align*}
$$

**2. 足し算 (Addition)**

足し算の場合は単位が相殺されず、掛け算として蓄積されます。
$$
\begin{align*}
\ln A_1 + \ln A_2 &= \ln (\{A_1\}[A_{\cancel{1}}]) + \ln (\{A_2\}[A_{\cancel{2}}]) \\
&= \ln (\{A_1\} \cdot \{A_2\}) + \ln ([A] \cdot [A])
\end{align*}
$$

> **ポイント:**
> $\ln [A]$ （単位の対数）という表現は直感的には少し不思議に見えますが、熱力学（気体の体積変化など）の計算プロセスで実際に現れる数学的なテクニックです。最終的に引き算 **（差分）** をとることで次元が消え、結果的に「物理法則として矛盾しない」仕組みになっています。
