### 加速度

#### 速度が一定でない物体の変化

静止している物体が動き出すことを考えると、(瞬間の)速度は常に一定ではありません.  
そこで速度$v$ と時間$t$ が次のように変化する物体を考えます.

> 図1.6 のようなグラフを $v-t$ グラフ とよびます.

* 下の枠をクリックし, 再生ボタンを押してください.




```
#@title TEXT p5 図1.6 v-t グラフにおける加速度 { vertical-output: true }
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import numpy as np

(V,T)=(5.0,2.5)

plt.figure()
t = np.linspace(0, T)
plt.xlabel("$t$ [s]")
plt.ylabel("$v$ [m/s]")
plt.xlim(0, 5.0)
plt.ylim(0, 5.5)
a = V/T
plt.plot(t, a*t, color="blue")
plt.show()
```

### 加速度の定義

下図のように,

* 始めの状態が, 時間 $t_1$ で(瞬間の)速度 $v_1$
* あとの状態が, 時間 $t_2$ で(瞬間の)速度 $v_2$

とします.  これから

* 時間変化を $\varDelta t = t_2 - t_1$  
* 速度変化を $\varDelta v = v_2 - v_1$ 

として, 加速度を次のように定義します.

加速度は英語で acceleration なので, その頭文字 $a$ をとり
$$
(1.3)\qquad \bar{a} 
= \frac{\varDelta v}{\varDelta t}
= \frac{v_2 - v_1}{t_2 - t_1}
$$
とします.
これを記号を利用せずに単語で表現すれば
$$
(1.3)'\qquad \text{(平均の)加速度} 
= \frac{\text{速度変化}}{\text{時間変化}}
$$
となります. 
この式より単位だけ調べると
$\dfrac{\rm m/s}{\rm s}={\rm m/s^2}$
となり
> 加速度の単位は, 国際単位系では m/s<sup>2</sup> 

となります. 加速度の定義を文章で表現すると

> (平均の)加速度は, 単位時間あたりの速度変化で定義される

といえます.

#### 加速度の注意点

* 時間が経つことを考えているので, 時間変化 $\varDelta t=t_2 -t_1$ は常に正(+)
* $\varDelta v$ の正負(+,-)によって, (平均の)加速度の正負が決まる
