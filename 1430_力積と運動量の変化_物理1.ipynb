# 力積と運動量

## 力積と運動量の変化
\begin{align*}
\bbox[magenta,2pt]{m v_1} - \bbox[cyan,2pt]{m v_0} &= \bbox[yellow,2pt]{F \Delta t}&
\end{align*}

- 後の運動量:: $\bbox[magenta,2pt]{m v_1}$
- 始めの運動量:: $\bbox[cyan,2pt]{m v_0}$
- 途中に働く力積:: $\bbox[yellow,2pt]{F \Delta t}$
  - 物体が力を受ける時間 $\Delta t$

これを踏まえて左辺を読み解けば

> 物体の運動量変化 $(mv_1 - mv_0)$ は,  
> 物体が受けた力積 $F \Delta t$ に**等しい**



## 力積のグラフ

力積は
\begin{align*}
F \Delta t = F \cdot \Delta t
\end{align*}

となり
>> $力$ と $時間(変化)$ を作図すると, **力と時間軸で囲まれた面積** 

で表される.

### 力の作用の仕方

実際の力は,
1. 接触直後はじっくりと
2. 重心に近づくほどしっかり
3. 離れ間際にはじっくりと

と作用する.
