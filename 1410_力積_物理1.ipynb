# 力積と運動量

$\require{cancel}$
\begin{align*}
&m\cdot a = F,&  a = \dfrac{\Delta t}{\Delta t}&\\
&m \left( \dfrac{\Delta v}{\bcancel{\Delta t}}\right) \bcancel{\Delta t} = F \Delta t&\\
&m \Delta v = F \Delta t,&  \Delta v = v_1 -v_0&
\end{align*}

この考察から
\begin{align*}
& m v_1 - m v_0 = F\Delta t&
\end{align*}

$\require{cancel}$
## 力積
これまでの考察で得られた式で
- 測り始め $t_0=0$  
  -測り初めからの経過時間 $t$  
- $v_1 \to v'$
- $v_0 \to v$

として整理すると
\begin{align*}
& m v_1 - m v_\bcancel{0} = Ft_\bcancel{1} -F\cancelto{0}{t_0}&
\end{align*}

次の教科書 p35 (1.21) が得られる
\begin{align}
& m v' - m v = F t.&
\end{align}




この式 (1.21) で, 初速度$v=0$とすると

\begin{align}
m v' &= F t&
\end{align}



$m=2\ {\rm kg}$ として, 下記の表の条件 1,2 を考えてみます.
