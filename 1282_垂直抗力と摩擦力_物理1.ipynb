あらい水平な床の上に物体が置かれていることを考えます.
- 外力 $F_0$ を加えますが, 物体はいきなり動きません.
- これを実現するためには, 右向きの力が働かなければなりません.

この力 $f$ を静止させるために必要な力を
> **静止摩擦力**

 と呼びます

 力のつり合いで外力 $F_0$ と 静止摩擦力 $f$ の関係を見てみます.
 右向きを正として
 \begin{align*}
  F_0 + ( -f ) &= 0&\\
  F_0 &= f&
 \end{align*}
 となります.

この静止摩擦力は, 外力$F_0$の大きさに依存することがわかりましたが,   
外力を大きくしていくと物体は動き出します.

それまでは, 静止摩擦力は順調に増えていきますが, **最大**をもつはずです.   
そのような静止摩擦力を **最大静止摩擦力 $f_{\rm max}$**  といいます. 

この $f_{\rm max}$ だけは, 床との関わりをあらわす垂直抗力 $N$ と比例関係にあります.
\begin{align*}
f_{\rm max} &\propto N&\\
f_{\rm max } &= \mu N&
\end{align*}
ここで $\mu$ を静止摩擦係数と呼びます.
