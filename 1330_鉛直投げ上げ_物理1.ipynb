$\require{cancel}$

初速度 $v_0＞0$, 
座標は **鉛直上向きを正** とする。重力の大きさ$W=mg$(重力以外は無視)
\begin{align*}
m a &= - W &\\
\bcancel{m} a &= -\bcancel{m} g &\\
a &=  - g &
\end{align*}
物体の加速度は **質量に依存しない** ことがわかります.

$t$ 秒後の速度 $v$ と 変位 $z$ は, 初速度 $v_0>0$ であることを踏まえると,  
等加速度直線運動の関係を, 鉛直**上向き** を $z$軸方向を正の方向として
\begin{align*}
v&={v_0} + \cancelto{-g}{a} t&\\
\cancelto{z}{s}&={v_0} t + \dfrac{1}{2} \cancelto{-g}{a} t^2,&
\end{align*}
整理すると
\begin{align*}
v_{z}&= v_0 - g t&\\
z&=v_0 t - \dfrac{1}{2} g t^2,&
\end{align*}

```
#@title 鉛直投げ上げ運動の速度と変位の時間変化
%matplotlib inline
import matplotlib.pyplot as plt
from ipywidgets import interact
import numpy as np
plt.style.use('ggplot')

(T,g)=(10,9.8)

v0=9.8*2

plt.figure()
t = np.linspace(0, T)

#@interact(a=(-7,+7))
#def f(a):
fig = plt.figure(figsize=(8,4))

# グラフ間の幅を調整
plt.subplots_adjust(wspace=0.5, hspace=0.5)

at = fig.add_subplot(1, 2, 1)
at.plot(t , v0-g*t ,color='b')
at.plot(t , 0*t ,color='k')
at.set_xlabel('$t$ [s]') 
at.set_ylabel('$v_z$ [m/s]')
at.set_title('$v_z-t$')
at.set_xlim([0, 10/2])
at.set_ylim([-g*10/5, g*10/5])   

vt = fig.add_subplot(1, 2, 2)
vt.plot(t , v0*t-1/2*g*t**2 )
vt.set_xlabel('t [s]')
vt.set_ylabel('$z$ [m]')
vt.set_title('$z-t$')
vt.set_xlim([0, 10/2])
vt.set_ylim([0, 1/2*g*10**2/10/2])

plt.show()
```
