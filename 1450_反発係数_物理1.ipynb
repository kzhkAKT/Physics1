$\require{cancel}$
速度の向きに注意し, 上の3つの終状態について 反発係数(衝突前後の速さの比) $e$を求めてみましょう


1. 始状態 vs. 終状態:1,  
$\left| v\right| = \left| v'\right|$: そのまま跳ね返る
\begin{align*}
e_1 &= \dfrac{\left| v' \right|}{\left| v  \right|} = \dfrac{\bcancel{\left| v \right|}}{\bcancel{\left| v  \right|}} =1&\\
&= \dfrac{-v'}{v}=1&
\end{align*}

2. 始状態 vs. 終状態:2,  
$\left| v\right| > \left| v'\right|$: ちょっと勢いが減る
\begin{align*}
e_2 &= \dfrac{\left| v' \right|}{\left| v  \right|} 
< \dfrac{\bcancel{\left| v \right|}}{\bcancel{\left| v  \right|}}=1&\\
&= \dfrac{-v'}{v} < 1&
\end{align*}

3. 始状態 vs. 終状態:3,  
$\left| v'\right|=0$: 壁と一緒になる
\begin{align*}
e_3 &= \dfrac{\left| v' \right|}{\left| v  \right|} 
= \dfrac{0}{\left| v  \right|} =0&
\end{align*}

ここまでの考察で, 速度 $v, v'$ を用いれば 反発係数 $e$ は,
\begin{align*}
e &= \dfrac{-v'}{v} = \dfrac{物体が壁から遠ざかる速さ}{物体が壁に近づく速さ}&
\end{align*}
と表せることがわかり, さらに $e$ の範囲も
>> \begin{align*}
0 \leqq e \leqq 1
\end{align*}
となることもわかりました.

そこで 反発係数 $e$ の範囲を三つに分けて,   
それぞれの衝突に名称を一般的に次のように呼びます.

1. $\quad e=1$:: 弾性衝突
2. $\quad 0 < e < 1$:: 非弾性衝突
3. $\quad e=0$:: 完全非弾性衝突

上の図のような二物体の衝突では, 物体(2) を壁にする,つまり動かないようにすれば,
> 物体(1) が 物体(2) に近づく, または遠ざかる速さの比

として, 反発係数 $e$ を定義できます.

>>\begin{align*}
e &= \dfrac{物体(1)が物体(2)から遠ざかる速さ}{物体(1)が物体(2)に近づく速さ}&\\
&=\dfrac{-\left(v'_1-v'_2\right)}{v_1 -v_2}&\\
&= \dfrac{v'_2-v'_1}{v_1 -v_2}&
\end{align*}

もちろんこのときも
1. $\quad e=1$:: 弾性衝突
2. $\quad 0 < e < 1$:: 非弾性衝突
3. $\quad e=0$:: 完全非弾性衝突

となります.  特に 3 の条件では, 二つの物体は一体となって動きます.
