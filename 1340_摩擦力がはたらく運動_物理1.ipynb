### 重力 $W$の大きさ
\begin{align*}
W&=mg&
\end{align*}

### 垂直抗力の $N$ の大きさ
鉛直方向の力のつり合いを考える

\begin{align*}
&m \cdot 0 = N +(-W)&\\
N=W
\end{align*}

### 動摩擦力 $f'$ の大きさ
床との関わりを表す力との関係を整理する

\begin{align*}
f' & \propto N&\\
f' &= \mu' N
\end{align*}

## 動摩擦力が働く運動

話を簡潔にするために

1. 外力は $F$ で, 初速度 $v_0$ は **零でない**
1. 座標は初速度の向きを正

とします.

### 動摩擦力がない場合の運動方程式

\begin{align*}
m \cdot a_0 &= F&
\end{align*}

### 動摩擦力がを場合の運動方程式

\begin{align*}
m \cdot a &= F + (- f')&
\end{align*}
これまでの考察を加えると
\begin{align*}
m \cdot a &= F + (- \mu' mg),&
\end{align*}
これより加速度は
\begin{align*}
 a &= \dfrac{F}{m} - \mu' g&
\end{align*}
となり, 一次元では等加速度直線運動となります.

これより, $t=0$ の初速度を $v_0$ を利用すると $t>0$ 秒後の速度 $v$ と変位 $s$ は
\begin{align*}
v &= v_0 + at = v_0 + \left(\dfrac{F}{m} - \mu' g\right) t&\\
s&=v_0 t+ \dfrac{1}{2} at^2
= v_0 t + \dfrac{1}{2} \left(\dfrac{F}{m} - \mu' g \right)t^2&
\end{align*}
となります.

ただし, 変位 $s$ は $a<0$ のとき、動摩擦力は動いている時にしか生じないため、
$v\geqq 0$ までしか成立しません.

> Q. 上の式で $F＝0$ のとき, 停止するまでに必要な時間を求めてください.
